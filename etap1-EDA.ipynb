```python
def check_data(df, name, expected_rows, expected_columns):
    print(f"Sprawdzanie danych {name}")
    print(f"Liczba wierszy {len(df)}, oczekiwano {expected_rows}")
    print(f"Liczba kolumn {len(df.columns)}, oczekiwano {expected_columns}")
    
    # Sprawdzenie zakresu wartości pikseli
    pixel_stats = df.iloc[:, 1:].describe()  
    print("Zakres wartości pikseli")
    print(pixel_stats.loc[['min', 'max']])
    
    #  pierwsza kolumna to etykieta
    if name == 'dice':
        labels_range = range(1, 7)
    elif name == 'mnist_train':
        labels_range = range(0, 10)
    else:  
        labels_range = range(0, 5)
    labels_stats = df.iloc[:, 0].value_counts().sort_index()  
    print("Rozklad etykiet")
    print(labels_stats)
    if not all(label in labels_range for label in df.iloc[:, 0].unique()):
        print("Etykiety poza oczekiwanym zakresem")
    else:
        print("Etykiety w oczekiwanym zakresie")
    
    print("\n")

# wywołanie
check_data(dice_df, 'dice', 60000, 785)
check_data(mnist_train_df, 'mnist_train', 60000, 785)

```

```python
def check_data(df, name, expected_rows, expected_columns):
    print(f"sprawdzanie danych {name}")
    print(f"liczba wierszy {len(df)} oczekiwano {expected_rows}")
    print(f"liczba kolumn {len(df.columns)} oczekiwano {expected_columns}")
    
    # zakres wartosci pikseli
    pixel_stats = df.iloc[:, 1:].describe()  # kolumna pierwsza to etykieta
    print("Zakres wartości pikseli")
    print(pixel_stats.loc[['min', 'max']])
    
    if name == 'dice':
        labels_range = range(1, 7)
    elif name == 'mnist_train':
        labels_range = range(0, 10)
    elif name == 'cifar_all':  # dostosowanie do cifarr
        labels_range = range(0, 10)  # cifar ma 10 kategorii
    else:
        raise ValueError("Nieznany zbior danych")

# przyklad
check_data(dice_df, 'dice', 60000, 785)
check_data(mnist_train_df, 'mnist_train', 60000, 785)
check_data(cifar_all_df, 'cifar_all', 50000, 785)  # dostosowanie do cifar
```
